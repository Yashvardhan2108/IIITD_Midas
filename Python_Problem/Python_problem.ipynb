## To Get Tweets from the Username "MidasIIITD"

```python
#To get tweets from the username
tweets=api.GetUserTimeline(screen_name="midasIIITD",include_rts=1,count=200,exclude_replies=1,)
#print(tweets)
```

# Below cells are for getting Text,Time,ID,Media,Likes,Retweets from the GetUserTimeline function we called above 

```python
tweets=[i.AsDict() for i in tweets]
tweetMain={}
```

```python
id=[]
tweetText=[]
tweetTime=[]
tweetRetweet=[]
tweetLikes=[]
tweetMedia=[]
tweetMedias=[]
tweetMain['tweets']=[]
k=0
i=0
for t in tweets:
    print(t)
    print()
    id.append(t['id'])
    #print(t['id'])
    #print(" ")
    tweetText.append(t['text'])
    #tweetMain['tweets'].append()
    tweetTime.append(t['created_at'])
    if 'favorite_count' in t:
        tweetLikes.append(t['favorite_count'])
    else:
        tweetLikes.append(0)
    if 'retweet_count' in t:
        tweetRetweet.append(t['retweet_count'])
    else:
        tweetRetweet.append(0)
    if 'media' in t:
        tweetMedia.append(t['media'])
        tweetMedias.append(len(tweetMedia[i]))
    else:
        tweetMedia.append(0)
        tweetMedias.append(0)
    i+=1
    
     

#print()
#print()
print(tweetMedias)
    
```

## Below cell appends everything into a dictionary called tweetMain

```python
for i in range(0,len(id)):
    tweetMain['tweets'].append({'ID':id[i],'Text':tweetText[i],'Time': tweetTime[i],'Likes':tweetLikes[i],'Retweets':tweetRetweet[i],'Media': tweetMedias[i]})
```

## Below Cells will convert the tweetMain dictionary into a JSON File.

```python
import json
```

```python
with open('tweetMain.txt','w') as out:
    json.dump(tweetMain,out)
   
```

```python
with open('tweetMain.txt','r')as json_file:
    tweetMain=json.load(json_file)
    x=PrettyTable()
    y=PrettyTable()
    x.field_names=['ID','Likes','Retweets','Media']
    y.field_name=['ID','Text']
    for t in tweetMain['tweets']:
        if(t['Media']==0):
            x.add_row([t['ID'],t['Likes'],t['Retweets'],"None"])
        else:
            x.add_row([t['ID'],t['Likes'],t['Retweets'],t['Media']])
            
        y.add_row([t['ID'],t['Text']])
        
# Used PrettyTable as we wanted the information in tabular format, Pandas can also be used where we can convert tweetMain to a df.
```
